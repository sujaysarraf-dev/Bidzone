1. Draw the line represented by the function y=x.

```python
import matplotlib.pyplot as plt
import numpy as np

# Create x values
x = np.linspace(-10, 10, 100)

# Define y = x
y = x

# Plot the line
plt.plot(x, y)

# Add labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Graph of y = x")

# Show grid
plt.grid()

# Display the graph
plt.show()
```

3. Plot 20 values of x and y from the following functions.

𝑓(𝑥)=5𝑥3+2𝑥−1

𝑓(𝑥)=−2𝑥3+𝑥2+100

𝑓(𝑥)=2𝜋𝑥+20

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 20 values of x
x = np.linspace(-10, 10, 20)

# Define functions
y1 = 5*x**3 + 2*x - 1
y2 = -2*x**3 + x**2 + 100
y3 = 2*np.pi*x + 20

# Plot all functions
plt.plot(x, y1, label="f(x)=5x^3+2x-1")
plt.plot(x, y2, label="f(x)=-2x^3+x^2+100")
plt.plot(x, y3, label="f(x)=2πx+20")

# Labels and title
plt.xlabel("x")
plt.ylabel("f(x)")
plt.title("Plot of Three Functions")

# Show legend and grid
plt.legend()
plt.grid()

# Display graph
plt.show()
```

4. Plot the trigonometric COSINE function in interval [−3𝜋,3𝜋], i.e., 𝑦 = cos (𝑥), −3𝜋 ≤ 𝑥 ≤3𝜋

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values from -3π to 3π
x = np.linspace(-3*np.pi, 3*np.pi, 100)

# Compute cosine values
y = np.cos(x)

# Plot the graph
plt.plot(x, y)

# Labels and title
plt.xlabel("x")
plt.ylabel("cos(x)")
plt.title("Graph of y = cos(x) from -3π to 3π")

# Grid
plt.grid()

# Show the plot
plt.show()
```

5. Plot a circle with center at (a, b) and radius r. The values of a=b=2, and r=4, using a while loop

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
a = 2
b = 2
r = 4

x_vals = []
y_vals = []

# Start from 0 to 2π
t = 0
step = 0.1

while t <= 2*np.pi:
    x = a + r * np.cos(t)
    y = b + r * np.sin(t)

    x_vals.append(x)
    y_vals.append(y)

    t += step

# Plot the circle
plt.plot(x_vals, y_vals)

# Mark the center
plt.scatter(a, b)

# Labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Circle with center (2,2) and radius 4")

# Equal scaling for proper circle shape
plt.axis('equal')
plt.grid()

plt.show()
```

6. Plot the above circle and the ellipse on the same axes. This is give an idea about the intersection points of the two conic sections

```python
import numpy as np
import matplotlib.pyplot as plt

# Circle parameters
a = 2
b = 2
r = 4

# Ellipse parameters (choose semi-major and semi-minor axes)
A = 6   # horizontal radius
B = 3   # vertical radius

# Lists for circle (using while loop)
xc = []
yc = []

t = 0
step = 0.05

while t <= 2*np.pi:
    x = a + r * np.cos(t)
    y = b + r * np.sin(t)
    xc.append(x)
    yc.append(y)
    t += step

# Ellipse using numpy
t_vals = np.linspace(0, 2*np.pi, 200)
xe = a + A * np.cos(t_vals)
ye = b + B * np.sin(t_vals)

# Plot both
plt.plot(xc, yc, label="Circle")
plt.plot(xe, ye, label="Ellipse")

# Mark center
plt.scatter(a, b)

# Labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Circle and Ellipse on Same Axes")

# Legend and grid
plt.legend()
plt.grid()
plt.axis('equal')

# Show plot
plt.show()
```

7. Plot a Parabola (𝑦 − 𝑝) = 4𝑎(𝑥 − 𝑞)^2 with vertex at (𝑝,𝑞). The values of p, q, and a can be chosen appropriately.

```python
import numpy as np
import matplotlib.pyplot as plt

# Given values
p = 1
q = 2
a = 0.5

# Generate x values
x = np.linspace(-5, 5, 100)

# Parabola equation
y = p + 4*a*(x - q)**2

# Plot the parabola
plt.plot(x, y, label="Parabola")

# Mark the vertex
plt.scatter(q, p)

# Labels and title
plt.xlabel("x")
plt.ylabel("y")
plt.title("Parabola with vertex at (1, 2)")

# Grid and legend
plt.grid()
plt.legend()

# Show plot
plt.show()
```

task 2: The following table shows the different vehicles and the distance covered by them in the span of 5 days. Display the following table in line plot.

a. Plot “days” in X- axis and “distance covered” in Y- axis.

b. Title of the table “bike details in Plot”

```python
import matplotlib.pyplot as plt

days = ['Day 1', 'Day 2', 'Day 3', 'Day 4', 'Day 5']

enfield = [50, 40, 70, 80, 20]
honda   = [80, 20, 20, 50, 60]
yamaha  = [70, 20, 60, 40, 60]
ktm     = [40, 20, 30, 70, 60]

plt.figure(figsize=(10, 6))

plt.plot(days, enfield, marker='o', label='Enfield', linewidth=2)
plt.plot(days, honda,   marker='s', label='Honda',   linewidth=2)
plt.plot(days, yamaha,  marker='^', label='Yamaha',  linewidth=2)
plt.plot(days, ktm,     marker='D', label='KTM',     linewidth=2)

plt.title('Bike Details Plot', fontsize=16, fontweight='bold')
plt.xlabel('Days', fontsize=13)
plt.ylabel('Distance Covered in KMs', fontsize=13)
plt.xticks(fontsize=11)
plt.yticks(fontsize=11)
plt.legend(fontsize=11)
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
